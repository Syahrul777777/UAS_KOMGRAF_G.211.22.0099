```python
import pandas as pd

# Membaca file CSV
df = pd.read_csv('stock.csv')

# Menampilkan data
print(df)
```

Jawaban NO. 2 A

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.figure(figsize =(15,5))
plt.hist(df['Purchase Price'],bins=23, edgecolor='purple', color='aqua')


# Menambahkan label pada sumbu x dan y
plt.xlabel('Purchase Price')
plt.ylabel('Frekuensi')

# Menambahkan judul
plt.title('Histogram Purchase Price')

# Menampilkan histogram
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.figure(figsize =(15,5))
plt.hist(df['Current Price'],bins=25, edgecolor='black', color='pink')


# Menambahkan label pada sumbu x dan y
plt.xlabel('Current Price')
plt.ylabel('Frekuensi')

# Menambahkan judul
plt.title('Current Price')

# Menampilkan histogram
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.figure(figsize =(15,5))
plt.hist(df['Gain/Loss'],bins=25, edgecolor='black', color='purple')


# Menambahkan label pada sumbu x dan y
plt.xlabel('Gain/Loss')
plt.ylabel('Frekuensi')

# Menambahkan judul
plt.title('Gain/Loss')

# Menampilkan histogram
plt.show()
```
